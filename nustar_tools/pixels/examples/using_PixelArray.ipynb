# Using the `PixelArray` and `RawPixelArray` classes

These classes are largely used as helper classes, but they contain some methods that can be useful for basic visualization.

```python
%load_ext autoreload
%autoreload 2
```

```python
from nustar_tools.pixels import PixelArray as pa
from nustar_tools.utils import utilities
%matplotlib inline
```

Define the relevant data files and read the data.
The data itself is provided to the class so that it is easier for the user to do the desired filtering on the data.

```python
evt_file = '/home/reed/Documents/research/nustar/data/20180529/80410201001/event_cl/nu80410201001A06_cl.evt'
hk_file = '/home/reed/Documents/research/nustar/data/20180529/80410201001/hk/nu80410201001A_fpm.hk'
time_range=('2018-05-29 16:15:00', '2018-05-29 16:30:00')

evt_data, hdr = utilities.get_event_data(evt_file, time_range)
```

Define the `PixelArray` object.

```python
parr = pa.PixelArray(evt_data, hdr)
```

The object can be used to make lightcurves.

```python
fig, ax = parr.plot_lightcurve(12, hk_file=hk_file)
```

The `RawPixelArray` class will organize the data by its raw (detector) pixels, so each detector is a 32 x 32 grid of pixels.
This is particularly useful for visualizing pointing shifts since the source can be observed to move across the detectors.

```python
rparr = pa.RawPixelArray(evt_data, hdr)
out = rparr.plot_det_counts()
```
