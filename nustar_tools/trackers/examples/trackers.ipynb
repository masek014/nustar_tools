# CoordinateTracker

Exploring the different trackers implemented for *NuSTAR* time series data.

```python
%load_ext autoreload
%autoreload 2
```

```python
import matplotlib.pyplot as plt
from nustar_tools.trackers import CoordinateTracker as ct
%matplotlib inline
```

## Input

```python
id_dir = '/home/reed/Documents/research/nustar/data/20180529/80410201001/'
fpm = 'A'
time_range = ct.get_observation_time(id_dir, fpm)
```

## Attitude

```python
att_tracker = ct.AttitudeTracker(id_dir)
att_tracker.read_data(time_range)
# att_tracker.convert_to_solar()
att_tracker.make_overview('Attitude Coordinates', b_add_fft=False)
plt.show()
```

## Optical Axis

```python
oa_tracker = ct.OpticalAxisTracker(id_dir, fpm)
oa_tracker.read_data(time_range)
oa_tracker.convert_to_solar()
oa_tracker.make_overview('Optical Axis Coordinates', b_add_fft=False)
plt.show()
```

## Centroid

```python
raw_tracker = ct.CentroidTracker(id_dir, fpm, 2, 'RAW')
raw_tracker.read_data(time_range)
raw_tracker.make_overview(b_add_fft=False)
plt.show()
```

## Aperture Stop

```python
as_tracker = ct.ApertureStopTracker(id_dir, fpm)
as_tracker.read_data(time_range)
as_tracker.make_overview('Aperture Stop Coordinates', b_add_fft=False)
plt.show()
```

## DET2 Aperture Stop

```python
det2as_tracker = ct.Det2ApertureStopTracker(id_dir, fpm)
det2as_tracker.read_data(time_range)
det2as_tracker.make_overview('Det2 Aperture Stop Coordinates', b_add_fft=False)
plt.show()
```

## DET1 Reference

```python
det1_tracker = ct.Det1Tracker(id_dir, fpm)
det1_tracker.read_data(time_range)
det1_tracker.make_overview('DET1 Reference Coordinates', b_add_fft=False)
plt.show()
```

## Mast Aspect

According to the *NuSTAR* software guide, the `T_FBOB` column is the translation from the focal plane to the optical plane.

```python
mast_tracker = ct.MastTracker(id_dir)
mast_tracker.read_data(time_range)
mast_tracker.make_overview('Mast Aspect Overview', False)
plt.show()
```
